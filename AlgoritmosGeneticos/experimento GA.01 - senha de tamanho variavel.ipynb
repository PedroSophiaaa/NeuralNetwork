```python
from string import printable
from funcoes import new_pop_passw_size, fitness_passw_size, tourn_sel_min, crossover_passw_size, mutation_passw, mutation_size
import random as rd
```

## Códigos e discussão



```python
pass_chars = printable[:-5]
```

```python
def create_password(n):
    password = ""
    for i in range(n):
        password += rd.choice(pass_chars)
    return password
```

```python
TAM_POP = 10
NUM_GEN = 1000
GEN_MAX = 10000
PM = 0.05 # Probabilidade de mutação
PM_SIZE = 0.2 # Probabilidade de mutação de tamanho
PC = 0.75 # Probabilidade de cruzamento
MIN_SIZE = 3 # Tamanho mínimo da senha
MAX_SIZE = 100 # Tamanho máximo da senha
PS = rd.randint(MIN_SIZE, MAX_SIZE) # Número de caracteres da senha

PASSWORD = create_password(PS)
print("Real password: '", PASSWORD, "'", sep='')
```

```python
# Teste com uma senha mais simples

PASSWORD = "pepe"
pass_chars = "abcdefghijklmnopqrstuvwxyz"
```

```python
gen = 1
pop = new_pop_passw_size(TAM_POP, pass_chars, PASSWORD, MIN_SIZE, MAX_SIZE)

min_fit = float('inf')

while min_fit != 0:
    if gen > GEN_MAX:
        break
    print("gen ", gen, ":", sep="")
    if gen > 1:
        pop = tourn_sel_min(pop, PASSWORD)

        pop_inds = [pop[i]["ind"] for i in pop]
        p1_list = pop_inds[0::2]
        p2_list = pop_inds[1::2]
        c = 1
        for p1, p2 in zip(p1_list, p2_list):
            filho1, filho2 = crossover_passw_size(p1, p2, PC)
            pop[c]["ind"] = filho1
            pop[c + 1]["ind"] = filho2
            c += 2
    for i in pop:
        curr_ind = pop[i]
        if gen > 1:
            curr_ind = mutation_passw(curr_ind, pass_chars, PM)
            curr_ind = mutation_size(curr_ind, PM_SIZE, pass_chars)
        curr_ind["fitness"] = fitness_passw_size(curr_ind["ind"], PASSWORD)
        print("\t", curr_ind)
    for i in pop:
        if pop[i]['fitness'] == 0:
            best_ind = pop[i]['ind']
            best_ind_tag = pop[i]['tag']
            min_fit = 0
            print(f'\nPassword found!\nGen: {gen}\nInd: {best_ind_tag}\nPassword: {best_ind}')
    gen += 1
    print()
```
